```python
def f(x):
    if x <= 0:
        return np.cos(np.cos(x))
    elif 0 < x <= 4:
        return 1 + np.sqrt(x)
    else:
        return 0

def g(x):
    if x >= 1:
        return -x
    else:
        return 0

x = np.linspace(-6, 6, 1000)

f_x = np.array([f(xi) for xi in x])
g_x = np.array([g(xi) for xi in x])
f_g_x = np.array([f(gi) for gi in g_x])

plt.figure(figsize=(10,6))
plt.plot(x, f_x, label='f(x)')
plt.plot(x, g_x, label='g(x)')
plt.plot(x, f_g_x, label='f(g(x))')
plt.axhline(0, color='black', linewidth=0.7)
plt.axvline(0, color='black', linewidth=0.7)
plt.grid(True)
plt.legend()
plt.title("Графики функций f(x), g(x), f(g(x)) на отрезке [-6,6]")
plt.show()
```
